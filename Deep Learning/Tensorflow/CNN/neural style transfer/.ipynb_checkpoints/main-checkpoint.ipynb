### 3) Helper function to convert image to array

```python
def load_image(image):
    image = plt.imread(image)
    img = tf.image.convert_image_dtype(image, tf.float32)
    img = tf.image.resize(img, [400, 400])
    img = img[tf.newaxis, :]
    return img
```

```python
content = load_image('images/content.jpg')
style = load_image('images/style.jpg')
```

### 4) Instantiate the vgg19 model

```python
vgg = VGG19(include_top=False, weights='imagenet')
vgg.trainable = False
```

### 5) Understanding the vgg19

```python
# layers name
for layer in vgg.layers:
    print(layer.name)
```

### 7) Defining Gram matrix

```python
def gram_matrix(tensor):
    temp = tensor
    temp = tf.squeeze(temp)
    fun = tf.reshape(temp, [temp.shape[2], temp.shape[0]*temp.shape[1]])
    result = tf.matmul(temp, temp, transpose_b=True)
    gram = tf.expand_dims(result, axis=0)
    return gram
```
